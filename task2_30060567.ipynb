### 3. Processing

Here on the files which are read are tokenized, converted to lowercase and tokens with length less than 3 are removed.

```python
#File is tokenized,normalised and tokens with length <3 are removed.
new_files = []
for file in files:
    tokens = [item for item in word_tokenize(" ".join(file).lower()) if len(item) >= 3]
    new_files.append(tokens)

print(new_files)
```

The given stopword list is read and the new line '\n' is been replaced with the space " ".

```python
#Reading stopword list and replacing '\n' with "" 
with open('C:/Users/Public/python/stopwords_en/stopwords_en.txt') as f:
    new_stop_words = []
    stop_words = f.readlines()
    for word in stop_words:
        new_stop_words.append(word.replace("\n",""))
    print(new_stop_words)
    
```

Stopwords are removed from the tokenized files and the file is stored in list3. 

```python
#Removing stopwords from resumes
list3 = []
for sublist in new_files:
    for word in sublist:
        if word not in new_stop_words:
            list3.append(word)
print(list3)
```

First 200 bigrams are listed. 

```python
from nltk.util import ngrams
#bigrams
bigram_measures = nltk.collocations.BigramAssocMeasures()
finder = nltk.collocations.BigramCollocationFinder.from_words(list(list3))
finder.nbest(bigram_measures.pmi, 50)
finder.apply_freq_filter(7)
finder.apply_ngram_filter(lambda w1, w2: w1 == w2 or (w1 in new_stop_words and w2 in list3)) 
bigrams = finder.nbest(bigram_measures.pmi, 200) # PMI filter (best 200)
print(len(bigrams))
print(bigrams)
```

```python
wod= []
for t in list3:
    if t not in wod:
        wod.append(t)
print(wod)
```

```python
#Indexing the tokens
finaldict = {}
for i in wod:
    F = wod.index(i)
    finaldict[i] = F
print(finaldict)
    
```

Finally the final dictionary is exported to vocab.txt file with bigrams in it.

```python
#Writing the tokens in vocab text file
final_vocab = open('30060567_vocab.txt','w')
for key, value in finaldict.items():
    val = key + ':' + str(value) + "\n"
    final_vocab.write(val)
    
st = " First meaningful 200 bigrams \n"

#writing the first 200 bigrams in vocab text file
final_vocab.write(st)
for bi in bigrams:
    bi1 = bi[0] + ',' + bi[1] + '\n'
    final_vocab.write(bi1)

final_vocab.close()
```
